>Neuroverkon opetus sensoridatalla ja suuntatiedolla


```python
import numpy as np
import keras
from keras import layers
import matplotlib.pyplot as plt
import pandas as pd
from keras.utils import to_categorical

# Model / data parameters
num_classes = 6 # outputti 6 neuronia
input_shape = (3,) #inputti 3 neuronia

df = pd.read_csv('group5_data.csv',sep=';')

#Koko data
xyz_data = df.iloc[:,6:9].to_numpy() #xyz arvot
suunta = df.iloc[:,5].to_numpy() #suunta
# Muutetaan suunta kategoriseksi
suunta = to_categorical(suunta)

#Jaetaan data sekä suunta opetus ja testi dataan
x_train = xyz_data[113:,:] #heilutettua dataa
y_train = suunta[113:,:]

x_test = xyz_data[0:112,:] #vakaa data
y_test = suunta[0:112,:]

#print(xyz_data[0:10,:])
#print(suunta[0:30])
#print('xyz: ' + str(xyz_data.shape))
#print('suunta: ' + str(suunta.shape))

#Mallin rakennus
model = keras.Sequential(
    [
        keras.Input(shape=input_shape), #input layer
        layers.Dense(10,activation='relu'), #hidden layer
        layers.Dense(num_classes, activation="softmax"), #output layer
    ]
)

model.summary()


#Mallin opetus
batch_size = 100
epochs = 1000

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs) #, validation_split=0.1)#tämä sekoitti tarkkuutta

```

```python
# Testataan mallin suorituskykyä testidatalla
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)

print('\nTest accuracy:', test_acc)
```

```python
# Ennustetaan testidatan luokat
predictions = model.predict(x_test)

# Muutetaan ennusteet luokiksi
ennustettu = np.argmax(predictions, axis=1)
todellinen = np.argmax(y_test, axis=1)

# Tulostetaan ensimmäiset 10 ennustetta ja todellista luokkaa
for i in range(100):
    print(f"Ennuste: {ennustettu[i]}, Todellinen luokka: {todellinen[i]}")
```
